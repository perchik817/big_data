### Расширенные возможности построения графиков
Следующий код строит несколько графиков в одном окне и сохраняет результат в графический файл .png


````python
# independed var
x=np.linspace(-4*np.pi, 4*np.pi, 100)
xlabel ="time sec"
ylabel="signal amplitude"

# first graph
y1=np.sin(2*x)*np.exp(-0.1*x)
leg1="e(-0.1*t)*sin(2*t)"

# second graph
y2=10*np.sin(0.5*x)+np.cos(10*x)
leg2="mixed signal"

# third graph
y3=np.sin(x)
leg3="reference signal"

# grid on x- and y-axises
plt.grid(color="g", linewidth=1)
plt.xlabel(xlabel, fontsize="x-large")
plt.ylabel(ylabel, fontsize="x-large")
plt.title("Harmonic signals")

plt.plot(x, y1, "y--", label=leg1)
plt.plot(x, y2, "p:", label=leg2)
plt.plot(x, y3, "b-", label=leg3)

plt.legend(loc="best")
plt.grid(True)
plt.savefig("3signals.png", format="png")
````

###Два графика на диаграмме в индивидуальных масштабах

Если nr*nc<10, то передавать параметры nr,nc,np можно без

###figure() и axes()

>команда ```axes()``` позволяет определить положение осей как ```axes([left, bottom, width, height])```, где все значения изменяются от 0 до 1.


````python
def f(t):
  return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.axes([0.55, 0.55, 0.4, 0.4])
plt.plot(t1, f(t1), "bo", t2, f(t2), "k")
plt.axes([0.3, 0.3, 0.4, 0.4])
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
plt.axes([0.05, 0.05, 0.4, 0.4])
plt.plot(t1, f(t1), "bo", t2, f(t2), "k")

plt.show()
````

###Перенос координатных осей в центр графика

* Функция ```plt.gca()``` возвращает текущие оси, с которыми мы можем взаимодействовать.
*```ax.spines['left']```: Это левая вертикальная ось. Мы устанавливаем её положение на ```'center'```, что означает, что она будет перемещена в центр графика, где **X=0**.
*```ax.spines['bottom']```: Это нижняя горизонтальная ось. Мы также перемещаем её в центр, где **Y=0**.

````python
X = np.linspace(-2,2,100)
Y = X**3

plt.plot(X, Y)
ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
````

Пример переноса осей для **subplot**:

````python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(X, Y)
ax1.set_title('Sharing Y axis')

ax1.spines['left'].set_position('center')
ax1.spines['bottom'].set_position('center')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

ax2.scatter(X, Y)
ax2.spines['left'].set_position('center')
ax2.spines['bottom'].set_position('center')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

f, axarr = plt.subplots(2, 2) # Four axes,
                               # returned as a 2-d array
axarr[0, 0].plot(X, Y)
axarr[0, 0].spines['left'].set_position('center')
axarr[0, 0].spines['bottom'].set_position('center')
axarr[0, 0].spines['top'].set_visible(False)
axarr[0, 0].spines['right'].set_visible(False)

axarr[0, 0].set_title('Axis [0,0]')
axarr[0, 1].scatter(X, Y)
axarr[0, 1].set_title('Axis [0,1]')
axarr[1, 0].plot(X, X ** 2)
axarr[1, 0].set_title('Axis [1,0]')
axarr[1, 1].scatter(X, X ** 1)
axarr[1, 1].set_title('Axis [1,1]')

plt.show()
````

### График рассеяния

График рассеяния (scatter plot) - это визуализация, которая позволяет сравнивать два набора данных, представляя их как множество точек на плоскости. Каждая точка имеет две координаты, соответствующие значениям из двух наборов данных.

Зачем использовать:
>Определение связи: Позволяет визуально оценить, есть ли какая-либо зависимость между двумя переменными.
Анализ распределения данных: Показывает, как данные распределены в пространстве, выявляет выбросы и кластеры.
Определение пределов измерений: Помогает установить минимальные и максимальные значения для данных.

Функция: ```matplotlib.pyplot.scatter()```
Основные параметры:

    x, y: Последовательности значений для осей X и Y соответственно.
    s: Размер маркеров (может быть числом или массивом).
    c: Цвет маркеров (может быть цветом или массивом цветов).
    marker: Тип маркера (круг, квадрат, треугольник и т.д.).

````python
rg = np.random.default_rng(5)

means = (0.5, 0.9)
covar = [[1., 0.6], [0.6, 1.]]
data = rg.multivariate_normal(means, covar, 5000)

x = np.random.randn(1000)
y = np.random.randn(1000)
size = 5
colors = np.random.rand(1000)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].scatter(data[:, 0], data[:, 1], marker='o', s=1)
axs[0].set_title("Scatter Plot")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")

axs[1].scatter(x, y, s=size, c=colors)
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")

plt.tight_layout()

plt.show()
````

### Экспериментальные данные

Экспериментальные точки с усами и теоретическая кривая на одном графике.
plt.plot(xt,yt)
plt.errorbar(xe,ye,fmt='ro',yerr=yerr)


график, созданный с помощью plt.errorbar(xe, ye, fmt='ro', yerr=yerr), отображает экспериментальные данные, полученные в результате измерений функции 1 / (x^2 + 1) в диапазоне x от -3 до 3. Эти данные представляют собой 21 точку с ошибкой, равной 0.1. Ошибки отображаются в виде вертикальных линий, соединяющих точки с ошибками.

````python
xt=np.linspace(-4,4,101)
yt=1/(xt**2+1)
xe=np.linspace(-3,3,21)
yerr=0.1*np.ones(21)
ye=1/(xe**2+1)+yerr*np.random.normal(size=21)
plt.plot(xt,yt)
plt.errorbar(xe,ye,fmt='ro',yerr=yerr)
plt.grid(True)
plt.show()
````

### Гистограмма

Сгенерируем N случайных чисел с нормальным (гауссовым) распределением (
среднее 0, среднеквадратичное отклонение 1), и раскидаем их по 20 битам от −3 до 3 (точки за пределами этого интервала отбрасываются). Для сравнения, вместе с
гистограммой нарисуем Гауссову кривую в том же масштабе. И даже напишем
формулу Гаусса.

### Построение 3D графиков. Работа с mplot3d Toolkit

#### Линейный график

Axes3D.plot:

* xs:
      Тип: одномерный массив (1D array).
      Описание: значения координат по оси X.

* ys:
      Тип: одномерный массив (1D array).
      Описание: значения координат по оси Y.

* zs:
      Тип: скаляр или одномерный массив (1D array).
      Описание: значения координат по оси Z. Если передан скаляр (одно значение), то оно будет применено ко всем точкам графика, то есть все точки будут иметь одно и то же значение по оси Z.

* zdir:
      Тип: строка, может принимать значения 'x', 'y' или 'z'.
      Значение по умолчанию: 'z'.
      Описание: определяет ось, которая будет считаться направлением Z. По умолчанию ось Z остаётся стандартным направлением, но если требуется поменять ось, можно указать 'x' или 'y'.

* kwargs:
      Описание: дополнительные аргументы, которые могут быть переданы функции. Они такие же, как и в функции plot() для построения двумерных графиков. Это могут быть параметры для настройки стиля графика, такие как цвет (color), тип линии (linestyle), толщина линии (linewidth) и другие.

````python
x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.cos(x)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve')
plt.show()
````

#### Каркасная поверхность

Axes3D.plot_wireframe():

* X, Y, Z:
      Тип: двумерные массивы (2D arrays).
      Описание: данные для построения каркасной поверхности. Массивы X и Y задают координаты на плоскости, а массив Z — это значения функции на этих координатах.

* rcount, ccount:
      Тип: int.
      Значение по умолчанию: 50.
      Описание: задаёт максимальное количество элементов каркаса, которое будет использовано для построения в каждом направлении — строки и столбцы. Это ограничение применяется для улучшения производительности при визуализации.

* rstride, cstride:
      Тип: int.
      Описание: параметры, задающие шаг, с которым будут выбираться элементы из строк и столбцов массивов X, Y, Z для построения каркасной поверхности. Чем больше шаг, тем реже будут выбираться элементы, что влияет на плотность линий каркаса.

* kwargs:
      Описание: дополнительные параметры, которые могут быть переданы через **kwargs. Они используются для управления стилем и параметрами линий каркасной поверхности через Line3DCollection. Это может включать цвет (color), толщину линий (linewidth), стиль линий и другие настройки.

````python
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)
ax.legend()
plt.show()
````
