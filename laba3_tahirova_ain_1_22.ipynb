### 2.   Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры



>subplot(122)
, то 2 колонки означает создание сетки для размещения нескольких графиков, в которой будет 2 столбца (то есть 2 графика будут размещены по горизонтали).

````python
import numpy as np
import matplotlib.pyplot as plt

# Генерация данных для квадратичной функции y = ax^2 + bx + c
a = 1
b = -2
c = 1

# Создаём массив значений x от -8 до 10 с шагом 0.1
x_quad = np.linspace(-8, 10, 400)
y_quad = a * x_quad**2 + b * x_quad + c

plt.plot(x_quad, y_quad, label=r'$y = x^2 - 2x + 1$')
plt.title("Graph of a Quadratic Function", fontsize=14, color="g")
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)

plt.grid(True)

plt.legend()

plt.show()
````

### 3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры


>Гармоническая функция — это решение уравнения Лапласа и обычно имеет вид синусоидальной или косинусоидальной формы, например, sin⁡(x)sin(x) или cos⁡(x)cos(x). Простейшими примерами гармонических функций являются синус и косинус, так как они описывают колебания и обладают свойством периодичности, что является ключевым для гармонических функций.

````python
x_harmonic = np.linspace(-2 * np.pi, 2 * np.pi, 300)
y_harmonic = np.cos(x_harmonic)

plt.plot(x_harmonic, y_harmonic)
plt.title("Graph of a harmonic function", fontsize=14, color='g')
plt.xlabel("X")
plt.ylabel("cos(X)")

plt.grid(True)

# Устанавливаем масштаб осей
plt.ylim(-1.2, 1.2)

plt.show()
````

### 4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

>Функция ```plt.tight_layout()``` в Matplotlib автоматически корректирует параметры расположения графиков на фигуре, чтобы они не перекрывались и выглядели более аккуратно. Это особенно полезно, когда на фигуре размещено несколько графиков, подписи осей или заголовки могут пересекаться. Matplotlib старается минимизировать пересечения между элементами графиков, такими как оси, метки и заголовки.

````python
figure = plt.figure(figsize=(13, 5))

plt.subplot(121)
plt.plot(x_quad, y_quad)
plt.title("Quadratic Function y = x^2 - 2x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(122)
plt.plot(x_harmonic, y_harmonic)
plt.title("Harmonic Function cos(x)")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.grid()

plt.tight_layout()
plt.show()
````

### 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

````python
plt.figure(figsize=(9, 5))
plt.plot(x_quad, y_quad, label="y = x^2 - 2x + 1", color="b")
plt.plot(x_harmonic, y_harmonic, label="y = cos(x)", color="r")
plt.xlabel("X-axis", fontsize=10)
plt.ylabel("Y-axis", fontsize=10)
plt.title("Graphs of Quadratic and Harmonic Functions")
plt.ylim(-1.1, 1.1)
plt.grid()
plt.legend()
plt.show()
````

### 6.	Построить график функции, заданной параметрически, $$x(t)=cos(3t), y=sin(2t), t \in [0, 4\pi]$$

````python
t = np.linspace(0, 4 * np.pi, 400)

x = np.cos(3 * t)
y = np.sin(2 * t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label = r"$x(t) = \cos(3t), y = \sin(2t)$", color="g")

plt.xlabel(r"$x(t) = \cos(3t)$", fontsize=12)
plt.ylabel(r"$y = \sin(2t)$", fontsize=12)
plt.title(r"Parametric Graph: $t \in [0, 4\pi]$", fontsize=14)

# Легенда
plt.legend()

# Показываем график
plt.grid(True)
plt.show()
````

### 9.	Построить график произвольной каркасной поверхности

>Команда ```X, Y = np.meshgrid(X, Y)``` используется для создания двумерных координатных сеток на основе одномерных массивов;  принимает два одномерных массива (или более) и создает два (или более) двумерных массива, представляющих все возможные комбинации координат.

````python
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-8, 8, 88)
Y = np.linspace(-8, 8, 88)
X, Y = np.meshgrid(X, Y)

# Определяем произвольную функцию для Z
Z = np.sin(np.sqrt(X**2 + Y**2)) - np.cos(X)

# Создаем фигуру и оси 3D
fig = plt.figure(figsize=(15, 7))
ax = fig.add_subplot(111, projection='3d')

# Построение каркасной поверхности
ax.plot_wireframe(X, Y, Z, color="m", alpha=0.5)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')
ax.set_title('Wireframe Surface')

plt.show()
````
