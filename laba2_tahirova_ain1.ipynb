#####1.1.	Создайте одномерный массив my_array, в котором по порядку все числа до 100, кратные 5


````python
my_array = np.arange(5,101,5)
my_array
````

##### 2.1.	Из полученного массива получить массив с кратными 10 числами

````python
my_array = my_array[my_array % 10 == 0]
# my_array % 10 == 0 создаёт логический массив, где True соответствует элементам, кратным 10
# лог. массив исп-ся в качестве индексации => позволяет получить соотв-ие усл. эл-ы
my_array
````

##### 3.1.	Получить двумерный массив с двумя столбцами

 Синтаксис ```np.pad(array, (before, after), 'constant')``` означает:
* ```before``` — сколько элементов добавить в начало массива,
* ```after``` — сколько элементов добавить в конец (разница в длине между двумя массивами).

```np.column_stack((arr1, arr2))``` берёт элементы из arr1 и arr2 и складывает их по столбцам. Первый элемент arr1 будет стоять в первой строке первого столбца, первый элемент arr2 — в первой строке второго столбца, и так далее.

````python
two_columns = my_array.reshape(5, 2)
two_columns
````

##### 4.1.	Получите набор my_second_array, в котором все элементы меньше предыдущего исходного на 5




````python
my_second_array = my_array - 5
my_second_array
````

##### 5.1. Соедините массивы, полученные в п. 1.3. и 1.4., горизонтально. Используйте .T для транспонирования матрицы.

Чтобы соединить массивы, полученные в предыдущих шагах, горизонтально и использовать .T для транспонирования в NumPy это делается с помощью функции ```np.hstack()```, которая объединяет массивы по горизонтали, и метода ```.T``` для транспонирования.

````python
# Преобразую массивы в векторы-столбцы (одномерные массивы преобразуются в двумерные с одним столбцом), что необходимо для корректного объединения по горизонтали.
combined_array = np.hstack((two_columns,my_second_array.reshape(5,2)))

# Транспонирование матрицы, меняя местами строки и столбцы
transposed_array = combined_array.T

print("Горизонтально объединённый массив:")
print(combined_array)
print("\nТранспонированный массив:")
print(transposed_array)
````

##### 6.1.	Разделите 2 и 3 строку на 4 верхних строки my_array. Используйте .T для транспонирования матрицы

````python
devided = combined_array[1:3,:4] / my_array[3]
transported_devided = devided.T
transported_devided
````

##### 7.1.	Оставить только те значения, которые не изменились после деления

````python
# Оставляем только те элементы, результат деления которых равен 1
unchanged_values = devided == 1
indices = np.argwhere(unchanged_values)
# оригинальные значения из my_array
result = combined_array[1:3, :][indices[:, 0], indices[:, 1]]

print(f"Значения, которые не изменились после деления (равны 1): {result}")
````

##### 8.1. Снова получить трехстрочный двумерный массив и поменять местами первую и третью строки

метод ```vstack()``` (от англ. vertical stack) объед. массивы в вертикальном направлении

````python
my_third_array = my_array + 5
three_arrays = np.vstack((my_second_array, my_array, my_third_array))
print(f'initial three arrays: \n{three_arrays}\n')
three_arrays[[0,2]] = three_arrays[[2,0]]
print(f'switched first and third rows in three arrays: \n{three_arrays}')
````

##### 9.1.	Получить среднее, медиану и стандартное отклонение для каждой строки, используя команды np.mean(), np.median(), np.std().


````python
print(f'row  1: {three_arrays[[0]]}\n'
      f'average: {np.mean(three_arrays[[0]])}\n'
      f'median: {np.median(three_arrays[[0]])}\n'
      f'standard deviation: {np.std(three_arrays[[0]])}\n')

print(f'row  2: {three_arrays[[1]]}\n'
      f'average: {np.mean(three_arrays[[1]])}\n'
      f'median: {np.median(three_arrays[[1]])}\n'
      f'standard deviation: {np.std(three_arrays[[1]])}\n')

print(f'row  3: {three_arrays[[2]]}\n'
      f'average: {np.mean(three_arrays[[2]])}\n'
      f'median: {np.median(three_arrays[[2]])}\n'
      f'standard deviation: {np.std(three_arrays[[2]])}')
````

##### 10.1.	Сделать нормализацию на максимальное значение вдоль строк, поделив каждый элемент массива на max() вдоль строк.

Параметр ```axis=1``` указывает, что вычисление идет вдоль строк.


>**Нормализация**: делится каждый элемент строки на соответствующее максимальное значение этой строки.

````python
# Нахождение максимальных значений вдоль каждой строки (axis=1)
max_values = np.max(three_arrays, axis=1)

# Нормализация по строкам: деление каждого элемента строки на её максимальное значение
normalized_array = three_arrays / max_values[:, np.newaxis]

print("Исходный массив:")
print(three_arrays)

print("\nМаксимальные значения по строкам:")
print(max_values)

print("\nНормализованный массив:")
print(normalized_array)
````
